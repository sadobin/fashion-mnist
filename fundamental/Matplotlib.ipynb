```python
                                    ### 1. Linear function ###

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(11)
y = np.arange(11)

# x = range(11)
# y = range(11)

plt.figure( figsize=(5,5) )
plt.plot(x, y, 'r--', label="Linear function")
plt.legend(loc="upper center")
plt.grid(True)
plt.ylabel("Y-axis")
plt.xlabel("X-asix")
plt.axis( [0, 14, 0, 12] )                             # Define limitation for x and y axis
```

```python
                                    ### 2. Sinosuid function ###

import numpy as np
import matplotlib.pyplot as plt

f = np.arange(0, 5, .01)
y = np.cos(2* np.pi * f)

plt.figure(figsize=(10,10))
plt.plot(f, y, 'b--', label="Sinusoid")
plt.grid(True)
plt.legend(loc="upper center")
plt.ylabel("Sine function")
plt.xlabel("Freqeuncy")
plt.axis( [0, 5, -1.1, 1.1] )
```
