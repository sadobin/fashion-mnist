```python
"""
    Activation functions:
        
        1. Rectified Linear Unit (ReLU)
            Advantage:
                - Computationally efficient
                - Non-linear
            
            Disadvantage:
                - The Dying ReLU problem

        
        2. Sigmoid
            Advantage:
                - Smooth gradient
                - Output values bound between 0 and 1
        
            Disadvantage:
                - Vanishing gradient
                - Outputs not zero centered
                - Computationally expensive
        
        
        3. Hyperbolic Tangent
            Advantage:
                - Zero centered
                - Output values bound between -1 and 1
            Disadvantage:
                - Vanishing gradient
                - Computationally expensive
        
        
        4. Softmax
            Advantage:
                - Able to handle multiple classes
                - Useful for output neurons
            Disadvantage:
                - Computationally expensive
"""
print()
```

```python
#! /bin/python3.8


import numpy             as np
import matplotlib.pyplot as plt


class Activation_Functions:
    
    def __init__(self, inputs):
        
        self.inputs = inputs
        self.instantiate()
        self.calculate()
        
        
    def calculate(self):
        
        self.softmax_den = self.softmax_denominator()

        for i in range( self.inputs.size ):
            value = self.inputs[i]
            pos   = np.exp(  value )
            neg   = np.exp( -value )
            
            self.relu[i]    = max(0, value ) 
            self.sigmoid[i] = 1 / ( 1 + neg ) 
            self.tanh[i]    = ( pos - neg ) / ( pos + neg ) 
            self.softmax[i] = pos / self.softmax_den 
            
    
    def instantiate(self):
        self.relu    = np.full( self.inputs.size, 0, dtype=float )
        self.sigmoid = np.full( self.inputs.size, 0, dtype=float )
        self.tanh    = np.full( self.inputs.size, 0, dtype=float )
        self.softmax = np.full( self.inputs.size, 0, dtype=float )

    
    def softmax_denominator(self):
        return sum( np.exp( self.inputs[i] ) for i in range( self.inputs.size ) )
            
        
    def plot(self, x, y, lab):
        plt.figure(figsize=(7,7))
        plt.plot(x, y, "r--", label=f"{lab} function")
        plt.grid(True)
        plt.legend(loc="upper left")
        plt.xlabel("Inputs")
        plt.ylabel(lab)
        plt.axis([x[0], x[-1], min(y)-0.025, max(y)+0.025])
        plt.show()

        
    # Getter functions
    def get_relu(self):    return self.relu
    def get_sigmoid(self): return self.sigmoid
    def get_tanh(self):    return self.tanh
    def get_softmax(self): return self.softmax
    

    

inputs = np.arange(-10, 10.1, .1, dtype=float)  # [-10 -9.9 ... 9.9 10]

functions = Activation_Functions( inputs )

relu    = functions.get_relu()
sigmoid = functions.get_sigmoid()
tanh    = functions.get_tanh()
softmax = functions.get_softmax()

# print("\n******* RESULT ********")
# print(inputs)
# print(relu)
# print(sigmoid)
# print(tanh)
# print(softmax)

functions.plot( inputs, relu,    "ReLU"    )
functions.plot( inputs, sigmoid, "Sigmoid" )
functions.plot( inputs, tanh,    "Tanh"    )
functions.plot( inputs, softmax, "Softmax" )
```
