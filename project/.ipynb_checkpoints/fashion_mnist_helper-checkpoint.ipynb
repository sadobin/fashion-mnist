```python
import numpy             as np
import matplotlib.pyplot as plt

print(f"[+] Version: {np.__version__}")
```

```python
"""
    ##### EXAMPLE-1: FASHION MNIST DATA #####

"""

from tensorflow.keras.datasets import fashion_mnist


[train_images, train_labels], [test_images, test_labels] = fashion_mnist.load_data()

print(f"[+] train_images: {train_images.shape}")
print(f"[+] train_labels: {train_labels.shape}")
print(f"[+] test_images: {test_images.shape}")
print(f"[+] test_labels: {test_labels.shape}\n")


print(f"[+] train_labels[0]: {train_labels[1]}")
print(f"[+] train_images[0]: {train_images[1]}")
```

```python
"""
    ##### EXAMPLE-2: VISUALIZE TRAIN IMAGES #####
"""
def plot():
    
    plt.figure(figsize=(11, 13))
    
    for i in range(100):
        plt.subplot(10, 10, i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow( train_images[i] )
        plt.xlabel( f"Number: {train_labels[i]}" )
        
    plt.show()


plot()

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
"""
    ##### EXAMPLE-3: VISUALIZE PREPROCESSED TRAIN IMAGES #####
"""

# Preprocessing data
train_images = train_images / 255
test_images = test_images / 255


def plot():
    
    plt.figure(figsize=(11, 13))
    
    for i in range(100):
        plt.subplot(10, 10, i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow( train_images[i], cmap=plt.cm.binary )
        plt.xlabel( f"Number: {train_labels[i]}" )
        
    plt.show()


plot()
```
